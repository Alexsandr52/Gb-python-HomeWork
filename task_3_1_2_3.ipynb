```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [10,np.cos(np.linspace(0, np.pi) * np.pi / 180)]# R , cos(A)
y = [10,np.sin(np.linspace(0, np.pi) * np.pi / 180)]# R , sin(A)

x_d = x[0]*x[1]
y_d = y[0]*y[1]

plt.grid(True)
plt.scatter(x_d, y_d)



```

```python
a = 30 * np.pi / 180

X = (1 - 10) * np.cos(a) + (1 - 15) * np.sin(a)
Y = -(1 - 10) * np.sin(a) + (1 - 15) * np.cos(a)

theta = np.linspace(0, 2 * np.pi, 100)
r = 5
x = X + r * np.cos(theta)
y = Y + r * np.sin(theta)

plt.grid(True)
plt.plot(x, y)
```
