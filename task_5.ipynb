```python
import numpy as np
import matplotlib as plt
from pylab import * 
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
my_axis = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)
z = 6*x + 10*y 
my_axis.plot_wireframe(x, y, z)
my_axis.plot_wireframe(x, y - 10, z)
```

```python
fig = figure()
my_axis = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)
z = x**2
my_axis.plot_wireframe(x, y, z)
```

```python
fig = figure()
my_axis = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)
z = y**2/2 + x**2/2
my_axis.plot_wireframe(x, y, z)
```
